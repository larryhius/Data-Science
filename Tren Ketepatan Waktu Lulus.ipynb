```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data Mahasiswa.csv')
data
```

```python
# Mengonversi status kelulusan menjadi numerik (1 untuk Tepat, 0 untuk Terlambat)
data['STATUS KELULUSAN'] = data['STATUS KELULUSAN'].map({'TEPAT': 1, 'TERLAMBAT': 0})

# Membuat pivot table berdasarkan status mahasiswa dan jenis kelamin (Persentase kelulusan)
pivot_table = data.pivot_table(index='STATUS MAHASISWA', values='STATUS KELULUSAN', aggfunc='mean') * 100

pivot_table
```

```python
# Membuat heatmap untuk melihat tren ketepatan waktu kelulusan berdasarkan status mahasiswa 
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Heatmap Ketepatan Waktu Kelulusan Berdasarkan Status Mahasiswa')
plt.ylabel('Status Mahasiswa')
plt.show()
```

```python
# Visualisasi Countplot
plt.figure(figsize=(8,6))
sns.countplot(x='STATUS MAHASISWA', hue='STATUS KELULUSAN', data=data)
plt.title('Jumlah Mahasiswa Berdasarkan Status Mahasiswa dan Status Kelulusan')
plt.show()
```
