```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
p = 0.55
n = 2000
# Simulate 2000 times
x = bernoulli.rvs(p, size=n)

# Loop for the result of each toss
for idx, toss in enumerate(x, 1):
    print(f"Toss {idx} = {toss}")
```

```python
cumulative_mean = np.cumsum(x) / np.arange(1, n + 1)

# Print cumulative mean and the complement (1 - mean)
for idx, mean in enumerate(cumulative_mean, 1):
    complement = 1 - mean
    print(f"Toss {idx}\nMean (x1) = {mean:.4f}, Complement (x0) = {complement:.4f}\n")
```

```python
# Plot cumulative mean and the theoretical probability
plt.figure(figsize=(10, 6))
plt.plot(cumulative_mean, label='Cumulative Mean of Tosses', color='blue')

# Plot the theoretical probability line at 0.55
plt.axhline(y=p, color='red', linestyle='--', label=f'Theoretical Probability ({p})')

# Adding labels and title
plt.xlabel('Number of Tosses')
plt.ylabel('Cumulative Mean (x)')
plt.title('Cumulative Mean of Coin Tosses vs Theoretical Probability')
plt.legend()

# Show the plot
plt.show()
```
